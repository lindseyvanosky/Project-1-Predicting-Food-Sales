```python
#Importing the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import matplotlib as mpl

food_sales = pd.read_csv("/content/drive/MyDrive/Coding Dojo/Data Sets/sales_predictions.csv")
df = food_sales
df.head(10)
```

```python
# Count the duplicates
df.duplicated().sum()
```

```python
#Identify missing values 
df.isna().sum()
```

```python
#Deciding how to handle NaN
#Determining how many unique values are in column Item_Weight
df.nunique()
#Considering 415 is a very small percentage of 8523, we will change all null values to the most common weight
```

```python
df.info()
```

```python
most_common_weight = df["Item_Weight"].mode()
```

##Filling null values

```python
df["Item_Weight"].fillna(1, inplace = True)
df.head(30)
```

```python
df["Item_Weight"].replace(1, float(most_common_weight), inplace = True)
df.head(30)
```

```python
df.isna().sum()
```

```python
#For Outlet size, there are only 3 unique values - high, medium, small
#Considering a quarter of our data is missing values in this column, we cannot delete those rows
df
```

```python
#Considering "Medium" is the median of these values, we can set NaN to Medium introducing only minor errors 
df["Outlet_Size"].fillna("Medium", inplace = True)
df.head()
```

```python
#Yay, no more missing values
df.isna().sum()
```

#Find and fix any inconsistent categories of data


```python
df["Outlet_Size"].value_counts()
```

```python
df["Item_Fat_Content"].value_counts()
```

```python
#We have some abbreviations of these values, so we have to make them synonumous
df["Item_Fat_Content"] = df["Item_Fat_Content"].replace("LF","Low Fat")
df["Item_Fat_Content"] = df["Item_Fat_Content"].replace("low fat","Low Fat")
df["Item_Fat_Content"] = df["Item_Fat_Content"].replace("reg","Regular")
df.head(10)
```

###Boxplot of Outlet Sales
- In this visualisation we can see the spread of Item Sales by Outlet Type. While this isn't the prettiest graph, it give is a good idea of the ranges of sales for each outlet type as well as the mean, min, max and outliers.

```python
import warnings
warnings.filterwarnings("ignore")
df.boxplot(column = "Item_Outlet_Sales", by = "Outlet_Type", notch = True, vert = False); 
plt.style.use("seaborn-poster");
plt.title("Item Sales by Storetype");
plt.xlabel("Item Sales");
plt.grid(b=None);
```

###Heatmap 
- With this visualization we can see we have some interesting correlations such as:
- We see a strong negative correlation between Item Visibility and Item Sales, which makes sense.
- The correlation betweem Item MRP and Item Sales is non-existant, meaning that the highest priced items don't always make the most money. 

```python
corr = df.corr()
plt.figure(figsize = (15, 7))
sns.heatmap(corr, cmap = "coolwarm", annot = True);
```

##Bar Graph of Item Sales by Item Type
- In this visualization we can see which categories of food sell the most! 

```python
fig, ax = plt.subplots(figsize=(10, 5))
food_sum = df.groupby(["Item_Type"])["Item_Outlet_Sales"].sum().round(2)
food_sum.plot(kind = "barh", x = "Item_Outlet_Sales", y = "Item_Type", color = ("pink", "purple", "blue", "green", "yellow", "orange", "red", "cyan", "white", "magenta"));
plt.xticks(rotation = 45);
plt.style.use("dark_background")
price_fmt = mpl.ticker.StrMethodFormatter('${x:,.0f}');
ax.xaxis.set_major_formatter(price_fmt);
plt.xticks(fontweight = "medium")
plt.yticks(fontweight = "medium")
plt.tick_params(axis= "both", which = "major", labelsize = 14);
plt.title("Food Sales by Category", fontsize = 16, fontweight = "bold", pad = 40, loc = "center");
plt.xlabel("Item Sales", fontsize = 16, fontweight = "bold", labelpad = 40);
plt.ylabel("Item Category", fontsize = 16, fontweight = "bold", labelpad = 40);
```

##Line Graph of Total Sale by Outlet Type
- In this visualization we can see that Supermarket Type 1 has the highest sales. Stores with this classification currently bring in the most money. 

```python
import matplotlib.font_manager
fig, ax = plt.subplots(figsize=(10, 5))    
total_sales = df.groupby(["Outlet_Type"])["Item_Outlet_Sales"].sum().round(2)
total_sales.plot(kind = "line", x = "Outlet_Type", y = "Item_Outlet_Sales", color = "cyan", marker = ".", markersize = 20, lw = 3);
price_fmt = mpl.ticker.StrMethodFormatter('${x:,.0f}');
ax.yaxis.set_major_formatter(price_fmt);
plt.tick_params(axis= "both", which = "major", labelsize = 14);
plt.xticks(fontweight = "bold", fontfamily = "sans-serif")
plt.yticks(fontweight = "bold", fontfamily = "sans-serif")
plt.title("Total Sales by Outlet Type", fontsize = 16, fontweight = "bold", pad = 40, loc = "center", fontfamily = "sans-serif");
plt.xlabel("Outlet Type", fontsize = 16, fontweight = "bold", labelpad = 40, fontfamily = "sans-serif");
plt.ylabel("Total Sales", fontsize = 16, fontweight = "bold", labelpad = 40, fontfamily = "sans-serif");
```
